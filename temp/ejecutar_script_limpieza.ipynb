```python
import os
import pathlib

def cargar_y_unir_dfs(ruta_carpeta, nombre_base="coches_segunda_mano", extension="csv"):
    
    archivos = [f for f in os.listdir(ruta_carpeta) if f.startswith(nombre_base) and f.endswith(f".{extension}")]  
    dfs = []  

    for archivo in archivos:
        ruta_completa = os.path.join(ruta_carpeta, archivo)  
        try:
            df = pd.read_csv(ruta_completa)  
            dfs.append(df)  
        except Exception as e:
            print(f"Error al leer {archivo}: {e}")  

    if dfs:
        df_final = pd.concat(dfs, ignore_index=True)  
        return df_final
    else:
        print("No se encontraron archivos.")
        return None
    
script_dir = pathlib.Path().resolve()
ruta_carpeta = os.path.join(script_dir, "../data")
  
df_consolidado = cargar_y_unir_dfs(ruta_carpeta, nombre_base="coches_segunda_mano", extension="csv")

df_consolidado.to_csv(os.path.join(ruta_carpeta, "coches_consolidado.csv"), index=False)

```

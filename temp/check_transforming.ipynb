```python
import matplotlib.pyplot as plt

df["kilometraje"].plot(kind="hist", bins=50, color="purple", alpha=0.7)
plt.title("Histograma de Kilometraje")
plt.xlabel("Kilometraje")
plt.ylabel("Frecuencia")
plt.grid(axis='y', alpha=0.75) #añade una cuadricula al eje y
plt.show()
```

```python
df[df["kilometraje"]<100].drop(columns=["fecha_extraccion", "ruta_imagen", "comunidad", "referencia", "peninsula_y_baleares", "nombre_vendedor"], inplace=False)
```

```python
df[df["kilometraje"]<100].drop(columns=["fecha_extraccion", "ruta_imagen", "comunidad", "referencia", "peninsula_y_baleares", "nombre_vendedor"], inplace=False)["url"].tolist()
```

```python
df[df["kilometraje"]>375_000].drop(columns=["fecha_extraccion", "ruta_imagen", "comunidad", "referencia", "peninsula_y_baleares", "nombre_vendedor"], inplace=False)["url"].tolist()
```

```python
df["potencia"].plot(kind="hist", bins=50, color="purple", alpha=0.7)
plt.title("Histograma de potencia")
plt.xlabel("potencia")
plt.ylabel("Frecuencia")
plt.grid(axis='y', alpha=0.75) #añade una cuadricula al eje y
plt.show()
```
